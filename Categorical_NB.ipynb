```python
import pandas as pd  
import numpy as np 
import matplotlib.pyplot as plt
dataset=pd.read_csv("CKD.csv") # correct dataset
print("dataset available")
df = pd.DataFrame(dataset)
num_rows = len(df.index)
num_columns = len(df.columns)
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg', 'sg', 'sg', 'sg', 'rbc', 'pc',
       'pcc', 'ba', 'htn', 'dm', 'cad',
       'appet', 'pe', 'ane']]
dependent=dataset[['classification']]
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset=dataset.replace({True:1,False:0})
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent, test_size=0.30 )
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown="ignore",sparse_output=False)
x_train_en = encoder.fit_transform(x_train)
x_test_en = encoder.transform(x_test)
```

```python
from sklearn.naive_bayes import CategoricalNB
model= CategoricalNB()
model.fit(x_train_en, y_train)
#model.predict(x_test7)
```

```python
y_pred=model.predict(x_test_en)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix  
cm = confusion_matrix(y_test,y_pred) 
print("The result for Bernoulli Navies Bayes (Categorical)  is:'\n'", cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,y_pred)
print(clf_report)
```
