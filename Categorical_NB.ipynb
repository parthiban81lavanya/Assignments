```python
import pandas as pd  
import numpy as np 
import matplotlib.pyplot as plt
dataset=pd.read_csv("CKD.csv") # correct dataset
print("dataset available")
df = pd.DataFrame(dataset)
num_rows = len(df.index)
num_columns = len(df.columns)
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset=dataset.replace({True:1,False:0})
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
df = pd.DataFrame(dataset)

x_train7,x_test7,y_train7,y_test7 = train_test_split(independent,dependent, test_size=0.30 )

from sklearn.preprocessing import StandardScaler   #Standard input data
sc = StandardScaler()
x_train7 = sc.fit_transform(x_train7)
x_test7 = sc.transform(x_test7)

categorical_columns = df.select_dtypes(include=["object", "category"]).columns.tolist()

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1), categorical_columns)
    ]
)

model = make_pipeline(preprocessor, CategoricalNB())

model.fit(x_train7, y_train7)
#model.predict(x_test7)
```
