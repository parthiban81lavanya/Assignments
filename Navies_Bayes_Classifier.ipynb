```python
#Navies Bayes Classification

import pandas as pd  
import numpy as np 
import matplotlib.pyplot as plt
dataset=pd.read_csv("CKD.csv") # correct dataset
print("dataset available")
df = pd.DataFrame(dataset)
num_rows = len(df.index)
num_columns = len(df.columns)
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset=dataset.replace({True:1,False:0})
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
#dataset
```

```python
from sklearn.model_selection import train_test_split
x_train6,x_test6,y_train6,y_test6 = train_test_split(independent,dependent, test_size=0.3,random_state=0)
```

```python
x_train6.shape,x_test6.shape,y_train6.shape,y_test6.shape
```

```python
dataset["classification_yes"].value_counts()
```

```python
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()
classifier.fit(x_train6,y_train6)
print("Model Created")
classifier
y_pred6 = classifier.predict(x_test6)
from sklearn.metrics import confusion_matrix  
cm = confusion_matrix(y_test6,y_pred6) 
print("The result for Gaussian Navies Bayes  is:'\n'", cm)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test6,y_pred6)
print(clf_report)

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.naive_bayes import MultinomialNB

classifier = MultinomialNB()
classifier.fit(x_train6,y_train6)
print("Model Created")
classifier
y_pred6 = classifier.predict(x_test6)
from sklearn.metrics import confusion_matrix  
cm = confusion_matrix(y_test6,y_pred6) 
print("The result for Multinomial  Navies Bayes  is:'\n'", cm)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test6,y_pred6)
print(clf_report)
```

```python
from sklearn.naive_bayes import ComplementNB

classifier = ComplementNB()
classifier.fit(x_train6,y_train6)
print("Model Created")
classifier
y_pred6 = classifier.predict(x_test6)
from sklearn.metrics import confusion_matrix  
cm = confusion_matrix(y_test6,y_pred6) 
print("The result for Complement Navies Bayes  is:'\n'", cm)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test6,y_pred6)
print(clf_report)
```

```python
from sklearn.naive_bayes import BernoulliNB 

classifier = BernoulliNB ()
classifier.fit(x_train6,y_train6)
print("Model Created")
classifier
y_pred6 = classifier.predict(x_test6)
from sklearn.metrics import confusion_matrix  
cm = confusion_matrix(y_test6,y_pred6) 
print("The result for Bernoulli Navies Bayes  is:'\n'", cm)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test6,y_pred6)
print(clf_report)
```

```python
from sklearn.naive_bayes import CategoricalNB 
classifier = CategoricalNB()
classifier.fit(x_train6,y_train6)
print("Model Created")
classifier
```
