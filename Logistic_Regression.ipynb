```python
#Logistic Regression is a Classificaiton Algorithim , Binary classification, Sigmid curve of threshold 0.5
#Grid Search 
import pandas as pd  
import numpy as np 
import matplotlib.pyplot as plt
dataset=pd.read_csv("CKD.csv") # correct dataset
print("dataset available")
df = pd.DataFrame(dataset)
num_rows = len(df.index)
num_columns = len(df.columns)
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
dataset.columns
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset=dataset.replace({True:1,False:0})
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
x_train4,x_test4,y_train4,y_test4 = train_test_split(independent,dependent, test_size=0.3,random_state=0)
```

```python
dataset["classification_yes"].value_counts()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}# 'liblinear' supports both l1 and l2 penalties
logreg = LogisticRegression(max_iter=1000)
grid_search = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1)

#classifier = LRCV(random_state=0, solver='liblinear',cv=5)
#classifier = LRCV(random_state=0, solver='liblinear')
#classifier.fit(x_train4,y_train4)
grid_search.fit(x_train4,y_train4)
print("Model Created")
```

```python
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score: ", grid_search.best_score_)
```

```python
result = grid_search.cv_results_
grid_predictions =  grid_search.predict(x_test4)
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test4,grid_predictions) 
print("The result for Logistic Regression CV is:'\n'", cm)
```

```python
#result
table = pd.DataFrame.from_dict(result)
table
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test4,y_pred4)
```

```python
#F1 SCORE
from sklearn.metrics import f1_score
f1_macro =  f1_score(y_test4,grid_predictions,average='weighted')
print("The f1 score vaule for best parameter {} is:'\n'".format(grid_search.best_params_),f1_macro) 
```

```python
#roc_auc Area Under Curve  1 is the best 
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test4,grid_search.predict_proba(x_test4)[:,1]) 
```
